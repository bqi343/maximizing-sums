```
def maximize_constrained(kappa, ell_p, ell_q):
    def get_val(p, q):
        return (
            (1 - kappa) * 2 * q * (1 - q)
            + 2 * kappa * (1 - q) * (1 - np.exp(-p))
            + 2 * p * ell_p
            + 2 * q * ell_q
        )

    max_val = 0
    for q in np.linspace(0, 0.5, 501):  # TODO: replace this?
        if q == 1 or kappa == 0:
            p = 0
        else:
            p = max(-np.log(-ell_p / kappa / (1 - q)), 0)
        p = min(p, 0.5)
        max_val = max(max_val, get_val(p, q))
    return max_val
```

```

def csm_get_val_unconstrained_ell(beta, kappa, ell_p, ell_q):
    if kappa < 0 or kappa > 1 or ell_p > 0:
        return 1
    return maximize_constrained(kappa, ell_p, ell_q) - beta * (ell_p + ell_q)
```
